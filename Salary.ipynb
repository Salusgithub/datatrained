```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()



data=pd.read_csv("C:\\Users\\geopt\\Downloads\\Salary.csv", sep=',', encoding='latin-1')


data
```

```python
g = sns.pairplot(data)
g.fig.subplots_adjust(top= 0.9);
```
