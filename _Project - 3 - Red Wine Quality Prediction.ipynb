## Feature scaling:

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(df_new.drop(['check'],axis=1))
fx=sc.transform(df_new.drop(['check'],axis=1))
x=pd.DataFrame(fx,columns=df_new.drop(['check'],axis=1).columns)
x
```

```python
#check for biasing in data
print(x.skew())
```

```python
# Number of features having skewness more than 0.55 
print(len(x.skew().loc[x.skew()>0.55]))
```

```python
# remove biasing
for index in x.skew().index:
    if x.skew().loc[index] >0.55:
        x[index]=np.cbrt(x[index])
```

```python
#lets again check the skewness
print(x.skew())
print()
print(len(x.skew().loc[x.skew()>0.55]))# Number of features having skewness more than 0.55 
```

```python
#check if columns can be further reduced
from sklearn.decomposition import PCA
for i in range(2,10):
    pca = PCA(n_components=i)
    principalComponents = pca.fit_transform(x)
    plt.figure()
    plt.plot(np.cumsum(pca.explained_variance_ratio_))
    plt.xlabel("number of components")
    plt.ylabel('Variance(%)') #for each component
    plt.title('Explained Variance')
    plt.show()
    print(i,sum(pca.explained_variance_ratio_))
```

so, the above plotting describes that around 95% of the variance is being explained by 7 components.
