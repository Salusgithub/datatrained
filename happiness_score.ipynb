```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()

data=pd.read_csv("C:\\Users\\geopt\\Downloads\\happiness_score.csv", sep=',', encoding='latin-1')

data
```

```python
g = sns.pairplot(data)
g.fig.subplots_adjust(top= 0.9);
```

```python
econ_happiness = ['Happiness Score','Economy (GDP per Capita)']

econ_corr = data[econ_happiness]

econ_corr.corr()

```

```python

sns.regplot(data = econ_corr, x = 'Happiness Score', y = 'Economy (GDP per Capita)').set_title("Happiness score vs Economy")
```
